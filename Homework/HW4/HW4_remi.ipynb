```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.decomposition import RandomizedPCA
```

```python
from sklearn.datasets import fetch_olivetti_faces 
faces = fetch_olivetti_faces()
```

```python
faces
```

```python
from sklearn.decomposition import PCA
```

```python
faces.data.shape
# 400 datapoints, 4096 features each
```

```python
faces.target
```

```python
X = faces.data
```

```python
pca = PCA().fit(X)
```

```python
ratios = pca.explained_variance_ratio_

comp_id = range(1, len(ratios)+1)
fig = plt.figure(figsize=(14,6))
plt.plot(comp_id, ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.xlim([0,50]) # only display the first 50 ratios
print "First 2 components explain {0}% of variation".format(100*ratios[0:2].sum().round(3))
print "First 15 components explain {0}% of variation".format(100*ratios[0:15].sum().round(3))
```

```python
from sklearn.preprocessing import scale
```

```python
newX = scale(X, axis=0, with_mean=True, with_std=True, copy=True )
```

```python
newX
```

```python
pca = PCA().fit(newX)
ratios = pca.explained_variance_ratio_
comp_id = range(1, len(ratios)+1)
fig = plt.figure(figsize=(14,6))
plt.plot(comp_id, ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.xlim([0,50]) # only display the first 50 ratios
```

```python
randpca = RandomizedPCA().fit(X)
randpca_ratios = randpca.explained_variance_ratio_
comp_id = range(1, len(ratios)+1)
fig = plt.figure(figsize=(14,6))
plt.plot(comp_id, randpca_ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.xlim([0,50]) # only display the first 50 ratios
print "First 2 components explain {0}% of variation".format(100*randpca_ratios[0:2].sum().round(3))
print "First 15 components explain {0}% of variation".format(100*randpca_ratios[0:15].sum().round(3))
```

### 3. Plot the datapoints in the plane of the two principal components. Are any clusters visible?

```python
pca = PCA(n_components=2).fit(X)
pca_X = pca.transform(X)
```

```python
pca_X1 = pca_X[:, 0:1]
pca_X2 = pca_X[:, 1:2]
```

```python
cmap = matplotlib.colors.ListedColormap(np.random.rand(40,3))

for target in range(0, 40):
    plt.scatter(pca_X1[faces.target == target],
                pca_X2[faces.target == target],
                c=cmap.colors[target])
```

## 4. Implement k-means clustering, using sklearn, with k = number of people in the dataset. Does the clustering work? Does each cluster map correctly to one person?

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=40, init='k-means++', n_init=3, max_iter = 300, random_state=1)
km.fit(pca_X)
# centroids = km.cluster_centers_
predictions = km.predict(pca_X)
predictions
```

```python
fig, (ax1, ax2) = plt.subplots( 1,2, figsize=(14,7), sharex=True, sharey=True)

for target in range(0, 40):
    ax1.scatter(pca_X1[faces.target == target],
                pca_X2[faces.target == target],
                c=cmap.colors[target])
    
    ax2.scatter(pca_X1[predictions == target],
                pca_X2[predictions == target],
                c=cmap.colors[target])
```

### 5. Vary the number k between 1 and the number of datapoints, and plot the Silhouette coefficient as a function of k. What do you observe? What conclusions can you draw?

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances

k_values = range(2, len(pca_X))
silhouette_scores = []

for k in k_values:
    km = KMeans(n_clusters=k,  init='k-means++', n_init=1, max_iter = 300, random_state=1)
    km.fit(pca_X)
    labels = km.labels_
    silhouette_scores.append(metrics.silhouette_score(pca_X, labels, metric='euclidean'))

fig = plt.figure(figsize=(14,6))
plt.scatter(k_values, silhouette_scores)
plt.xlabel('K')
plt.ylabel('Silhouette score')
plt.xlim([0,400]) # only display the first 50 ratios
plt.ylim([0,0.5]) # only display the first 50 ratios
```
