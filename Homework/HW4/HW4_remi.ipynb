```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.decomposition import RandomizedPCA
from sklearn.cluster import KMeans
```

### 1. Implement PCA on the dataset, using the sklearn. What do the principal components tell? (is scaling necessary?)

```python
from sklearn.datasets import fetch_olivetti_faces 
faces = fetch_olivetti_faces()
```

```python
faces.data.shape
# 400 datapoints, 4096 features each
```

```python
faces.target
# each image is mapped to one of 40 people
```

```python
X = faces.data
pca = PCA().fit(X)
```

```python
def plot_scree_for_pca(pca):
    ratios = pca.explained_variance_ratio_
    comp_id = range(1, len(ratios)+1)
    plt.figure(figsize=(14,6))
    plt.plot(comp_id, ratios, 'ro-', linewidth=2)
    plt.title('Scree Plot')
    plt.xlabel('Principal Component')
    plt.ylabel('Eigenvalue')
    plt.xlim([0,50]) # only display the first 50 ratios
    print "First 2 components explain {0:.{1}f}% of variation".format(100*ratios[0:2].sum(), 1)
    print "First 15 components explain {0:.{1}f}% of variation".format(100*ratios[0:15].sum(), 1)
    
plot_scree_for_pca(pca)
```

```python
# Let's try scaling to see if it improves anything

from sklearn.preprocessing import scale
newX = scale(X, axis=0, with_mean=True, with_std=True, copy=True )
```

We're getting the exact same scree plot. Scaling features does not help in this case.

### 2. Change from PCA to RandomizedPCA. Do you obtain the same principal components? Why? Why not?

```python
randpca = RandomizedPCA().fit(X)
plot_scree_for_pca(pca)
```

### 3. Plot the datapoints in the plane of the two principal components. Are any clusters visible?

```python
# Let's limit our PCA to 2 principal components
pca = PCA(n_components=2).fit(X)
pca_X = pca.transform(X)
```

```python
pca_X1 = pca_X[:, 0:1]
pca_X2 = pca_X[:, 1:2]
```

```python
# Each person will have one color assigned
cmap = matplotlib.colors.ListedColormap(np.random.rand(40,3))

plt.figure(figsize=(13,8))
for target in range(0, 40):
    plt.scatter(pca_X1[faces.target == target],
                pca_X2[faces.target == target],
                c=cmap.colors[target])
```

Using only 2 components, we're not seeing clear clusters on the Plot

## 4. Implement k-means clustering, using sklearn, with k = number of people in the dataset. Does the clustering work? Does each cluster map correctly to one person?

```python
km = KMeans(n_clusters=40, init='k-means++', n_init=3, max_iter = 300, random_state=1)
km.fit(pca_X)
predictions = km.predict(pca_X)
```

```python
fig, (ax1, ax2) = plt.subplots( 1,2, figsize=(14,6), sharex=True, sharey=True)

for target in range(0, 40):
    ax1.scatter(pca_X1[faces.target == target],
                pca_X2[faces.target == target],
                c=cmap.colors[target])
    
    ax2.scatter(pca_X1[predictions == target],
                pca_X2[predictions == target],
                c=cmap.colors[target])
```

The clusters determined by K-Means are pretty but seem pretty inaccurate.

```python
# Let's quantify the clustering accuracy

from sklearn import metrics
metrics.adjusted_rand_score(faces.target, predictions)  
```

The way to validate the accuracy of a cluster is to look at each pair and make sure the same pairs are still associated (or not).

Only 14% of pairs of images are correctly matched in this case.

### 5. Vary the number k between 1 and the number of datapoints, and plot the Silhouette coefficient as a function of k. What do you observe? What conclusions can you draw?

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances

k_values = range(2, len(pca_X))
silhouette_scores = []

for k in k_values:
    km = KMeans(n_clusters=k,  init='k-means++', n_init=1, max_iter = 300, random_state=1)
    km.fit(pca_X)
    labels = km.labels_
    silhouette_scores.append(metrics.silhouette_score(pca_X, labels, metric='euclidean'))

fig = plt.figure(figsize=(14,6))
plt.scatter(k_values, silhouette_scores)
plt.xlabel('K')
plt.ylabel('Silhouette score')
plt.xlim([0,400]) # only display the first 50 ratios
plt.ylim([0,0.5]) # only display the first 50 ratios
```

With 400 clusters (the number of data points), the Silhouette score is perfect but useless.

K around 70-80 seems to be the best way to generate good clusters based on our 2 components. Since we have a dataset with target values, it would be more useful to plot the clustering accuracy based on various model inputs like varying K, or varying the number of components extracted from PCA.
