## 1. Classify the raw data using a linear SVM. Do you need to perform several binary classifications or does scikit-learn support multi-class classification with SVMs?

```python
from sklearn.cross_validation import KFold

def cross_validate(X, y, classifier, k_fold):
    "Scores classifier using kfold cross_validation"
    # derive a set of (random) training and testing indices
    k_fold_indices = KFold(len(X), n_folds=k_fold,
                           shuffle=True, random_state=0)

    k_score_total = 0
    # train and score classifier for each slice
    for train_slice, test_slice in k_fold_indices :
        model = classifier(X[train_slice],y[train_slice])
        k_score = model.score(X[test_slice], y[test_slice])
        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
# According to the sklearn documentation, this class supports multiclass which is handled
# according to a one-vs-the-rest scheme (ovr)

from sklearn.svm import SVC

svc_model = SVC(C=1.0, kernel="linear").fit(data[features_names].values, data[target_name].values)
svc_predictions = svc_model.predict(data[features_names].values)
1.0 * sum(svc_predictions == data[target_name].values) / len(data[target_name].values)
```

## 2. Cross validate the result

```python
svc_model = SVC(C=1.0, kernel="linear").fit
cross_validate(data[features_names].values, data[target_name].values, svc_model, 5)
```

```python
svc_model = SVC(C=1000.0, kernel="linear").fit
cross_validate(data[features_names].values, data[target_name].values, svc_model, 5)
```

## 3. Preprocess the data with a normalization step, using the tools explained here: http://scikit-learn.org/stable/modules/preprocessing.html#standardization-or- mean-removal-and-variance-scaling

```python
from sklearn import preprocessing

X_scaled = preprocessing.scale(data[features_names].values)
```

## 4. Repeat the classification performed in step 1 using a linear SVM and crossvalidate the result. Is it better or worse?

```python
svc_model = SVC(C=1.0, kernel="linear").fit
cross_validate(X_scaled, data[target_name], svc_model, 5)
```

## 5. Learn about pipelines here: http://scikit-learn.org/stable/modules/pipeline.html implement a pipeline that comprises: - a preprocessing step - a classification step and run the pipeline on the raw data (not normalized)

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([('scaler', preprocessing.StandardScaler()), 
                     ('svc', SVC(C=1.0, kernel="linear"))]).fit

cross_validate(data[features_names].values, data[target_name].values, pipeline, 5)
```

## 6. Try varying the value of C or the type of kernel. Do you get better results?

```python
from sklearn.svm import SVC

# possible kernels: ‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’

kernels = ['linear', 'poly', 'rbf', 'sigmoid']
Cs = [ 0.1, 0.3, 1, 3, 10 ]

for k in kernels:
    for c in Cs:
        pipeline = Pipeline([('scaler', preprocessing.StandardScaler()), 
                             ('svc', SVC(C=c, kernel=k))]).fit
        score = cross_validate(data[features_names].values, data[target_name].values, pipeline, 5)
        print "Kernel: {0} | C: {1} | Score: {2}".format(k, c, score)
```

## 7. Learn about grid search here: http://scikit-learn.org/stable/modules/grid_search.html and feed your pipeline classifier to the grid search. Explore a range of values for C, gamma and the type of kernel. Can you find an optimum value?

```python
from sklearn import grid_search

pipeline_param_grid = [
  {'svc__C': [ 1e-09, 3e-09, 1e-08, 3e-08, 1e-07, 3e-07, 1e-06, 3e-06, 1e-05, 3e-05, 1e-04, 3e-04, 0.0001, 0.0003, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 2, 3, 10, 30 ], 
   'svc__gamma': [ 1e-09, 3e-09, 1e-08, 3e-08, 1e-07, 3e-07, 1e-06, 3e-06, 1e-05, 3e-05, 1e-04, 3e-04, 0.0001, 0.0003, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30 ], 
   'svc__kernel': ['linear', 'poly', 'rbf']},
 ]

pipeline = Pipeline([('scaler', preprocessing.StandardScaler()), ('svc', SVC())])
clf = grid_search.GridSearchCV(pipeline, pipeline_param_grid, cv=5)
clf.fit(data[features_names].values, data[target_name].values)

print clf.best_params_
print clf.best_score_
```
