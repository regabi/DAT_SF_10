```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
student_logins = pd.io.parsers.read_csv('student_logins.csv')
```

Converts time columns to datetime type

```python
for column_name in ['session_start_time', 'session_end_time', 'student_account_created']:
    student_logins[column_name] = pd.to_datetime(student_logins[column_name])
```

Add feature "session_duration" which represents the duration of sessions in minutes.

```python
student_logins['session_duration'] = student_logins['session_end_time'] - student_logins['session_start_time']
student_logins['session_duration'] = student_logins['session_duration'].map(lambda d: d.astype('timedelta64[m]') / np.timedelta64(1, 'm'))
student_logins['session_duration'] = student_logins['session_duration'].map(lambda d: 1 if d < 1 else d)
```

```python
# exclude logins where session_duration are beyond 99th percentile
student_logins = student_logins[student_logins['session_duration'] < student_logins['session_duration'].quantile(0.99)]
```

Adds 5 features: a, c, e, g, m to represent the class as 1/0 flags

```python
class_data = pd.get_dummies(student_logins['class_id'])
student_logins = student_logins.join(class_data)
```

```python
# Let's sort the data set by student_id, class_id.
student_logins = student_logins.sort(['student_id', 'class_id', 'student_previous_class_logins'])
student_logins.head()
```

```python
# Loops through every row to add feature "student_class_avg_session_duration"
# It is the sum of previous sessions for the user in a particular class

row_count = 0
previous_student_id = 0
previous_class_id = 0
total_minutes_in_class = 0
total_sessions_in_class = 0
student_class_avg_session_duration = []

for df_index, row in student_logins.iterrows():
#     print df_index
    row_count += 1
#     if row_count > 150:
#         break
    
    if row_count == 1:
        previous_student_id = row['student_id']
        previous_class_id = row['class_id']

    user_changed = row['student_id'] != previous_student_id or row['class_id'] != previous_class_id

    if user_changed:
#         print("\nStudent/Class changed to: {0}/{1}".format(row['student_id'], row['class_id']))
        previous_student_id = row['student_id']
        previous_class_id = row['class_id']
        historical_minutes_per_session = 0
        total_minutes_in_class = 0
        total_sessions_in_class = 0
    else:
        if total_sessions_in_class == 0:
            historical_minutes_per_session = 0
        else:
            historical_minutes_per_session = total_minutes_in_class / total_sessions_in_class

    student_class_avg_session_duration.append(historical_minutes_per_session)
    
    total_minutes_in_class  += row['session_duration']
    total_sessions_in_class += 1

#     print("minutes: {0}; session: {1}; avg: {2}".format(total_minutes_in_class, total_sessions_in_class, historical_minutes_per_session))


student_logins['student_class_avg_session_duration'] = student_class_avg_session_duration
```
